# Exploratory Data Analysis

# Gender

```python
df.columns
```

```python
ax = sns.countplot(x='Gender', data=df)
```

```python
gender_grph = sns.countplot(x='Gender', data=df)

for bars in gender_grph.containers:
    gender_grph.bar_label(bars)
```

```python
ax.clear()
```

```python
df.groupby(["Gender"],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_gen = df.groupby(["Gender"],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot ( x = 'Gender', y = 'Amount', data = sales_gen)
```

```python
sales_gen = df.groupby(["Gender"],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sales_gen_graph = sns.barplot ( x = 'Gender', y = 'Amount', data = sales_gen)

for bars in sales_gen_graph.containers:
    sales_gen_graph.bar_label(bars)
```

*From the above I graphs it can be derive that most number of buyers are females and female has the higher purchasing power than male.*

# State

```python
df.columns
```

```python
# total number of ordersfrom top 10 states

sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)
sales_state
```

```python
sns.set(rc = {'figure.figsize':(18,5)})
sns.barplot(data = sales_state , x = 'State', y = 'Orders')
```

*Highest number of orders are done by Uttar Pradesh People*

```python
# total number of ordersfrom top 10 states

sales_amt_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)
sales_amt_state
```

```python
sns.set(rc = {'figure.figsize':(18,5)})
sns.barplot(data = sales_amt_state , x = 'State', y = 'Amount')
```

### From the above graphs it can be derive that most of the orders made by the people of state residing Uttar Pradesh followed by Maharashtra followed by Karnataka

# Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
prdct_cat = sns.countplot(data = df, x = 'Product_Category')

for bars in prdct_cat.containers:
    prdct_cat.bar_label(bars)
```

```python
prdct_cat_amt_graph = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data = prdct_cat_amt_graph, x = 'Product_Category', y= 'Amount')
```

```python
prdct_cat_amt_graph = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,7)})
prdct_cat_amt_graph=sns.barplot(data = prdct_cat_amt_graph, x = 'Product_Category', y= 'Amount')

for bars in prdct_cat_amt_graph.containers:
    prdct_cat_amt_graph.bar_label(bars)
```

*From above graphs we can see that most of the sold products are from Food followed by Clothing and then Electronics category*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
